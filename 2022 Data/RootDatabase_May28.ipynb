```python
# match each county by the distance with the cloest county within the region

# 'buffCounty'

import math
import collections
county['buffCounty'] = ''
county['low_dis'] = 0

M = len(county)
N = len(df_create)


for i in range(M): 
    x1 = county.iloc[i,2] # la
    x2 = county.iloc[i,3] # long
    min_dis = float('inf')

    for idx in range(N): # region tbale
        y1 = df_create.iloc[idx,3] # la
        y2 = df_create.iloc[idx,4] # long

        dis = math.dist([x1,x2],[y1,y2])
        #lookup[combine_ca.iloc[idx,0]] = dis

        if dis < min_dis and dis != 0:
            min_dis = dis
            buffCounty = df_create.iloc[idx,0]
            
    county.iloc[i,4] = buffCounty
    county.iloc[i,5] = min_dis
```
