## Download Data

```python
region_need = ['San Diego CA', 
               'Los Angeles CA', 
               'San Francisco-Oakland-San Jose CA', 
               'Austin TX', 
               'Dallas-Ft. Worth TX', 'San Antonio TX']
```

```python
calfresh_2022 = pd.read_csv('/Users/chensumei/Desktop/Ram Paper/2022 Data/GoogleTrend/calfresh_2022/calfresh_2022_01.csv')
```

```python
calfresh_2022 = calfresh_2022.reset_index()
calfresh_2022.rename(columns={'index':'Region','Category: All categories':'Search'}, inplace=True)
```

```python
calfresh_2022
```

```python
# transfer str to datetime 

from datetime import datetime

date_string = calfresh_2022.iloc[0,1][-8:-1]

calfresh_2022['datetime'] = datetime.strptime(date_string, '%m/%d/%y')
```

```python
idx = []

for i in range(len(calfresh_2022)):
    for j in range(len(region_need)):
        if calfresh_2022.iloc[i, 0] == region_need[j]:
            idx.append(i)
            
```

```python
idx
```

```python
calfresh_2022 = calfresh_2022.loc[(idx)]
```

```python
calfresh_2022.fillna(0, inplace=True)
```

```python
calfresh_2022['Year'] = calfresh_2022['datetime'].dt.year
calfresh_2022['Month'] = calfresh_2022['datetime'].dt.month
```

```python
calfresh_2022['Search'] = calfresh_2022['Search'].astype('int32')
```
