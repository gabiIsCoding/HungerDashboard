## Transfer csv to json
- id + state + county + year + month + snap application number

```python
col_need = ['State', 'County', 'Year', 'Month', 'SNAP_Applications', 'index']

df_new = df[col_need]
```

```python
df_new.head()
```

```python
# Transfer data to the dictionary

df_new = df_new.to_dict(orient='records')
```

```python
import json

json_file = 'databaseForFunction.json'

with open(json_file, 'w') as file:
    json.dump(df_new, file)
```

```python
# review json file

with open(json_file, 'r') as file:
    data = json.load(file)

# Dump the data to a formatted JSON string
formatted_json = json.dumps(data, indent=4, sort_keys=True)

print(formatted_json)
```
