## Add one more feature - dummy value for state

```python
# split CA and texas using dummy

# create dummy variables
dummies = pd.get_dummies(df['State'], prefix='State')

# concatenate the dummy variables with the original dataframe
df = pd.concat([df, dummies], axis=1)
```

```python
# drop State_Texas since using dummy

df = df.drop(columns='State_Texas')
```

## Add one more feature - difference ratio for snap - May 2, 2023

```python
df.head()
```

```python
# drop rows whose last_snap is 0

# Drop rows with 0 in column'snap_change_ratio'
df = df.drop(df.loc[(df['last_snap'] == 0)].index)
```

```python
df['snap_change_ratio'] = round((df.SNAP_Applications / df.last_snap),3)
```

```python
df.head()
```

```python
# calculate the std of snap_change_ratio

df.snap_change_ratio.describe()
```

### Create frequency table

```python
snap = df[['snap_change_ratio']]
iw = (snap.max()-snap.min())/20
Abs_frecuency, intervals = np.histogram(snap, bins = 20)


# Create dataframe
df1 = pd.DataFrame(index = np.linspace(0,10,20), columns = ['start', 'end', 'class_marks','Frec_abs'])
# Assign the intervals
df1['start'] = intervals[:-1]
df1['end'] = intervals[1:]
# Calculate class marks
df1['class_marks'] = (df1['start'] + df1['end'])/2
# Assing Absolute frecuency
df1['Frec_abs'] = Abs_frecuency
```

```python
df1
```

```python
# Create bar plot
plt.bar(df1['class_marks'], df1['Frec_abs'], width = iw, color='blue', ec= 'black',alpha = 0.5)
# Add labels and modify size
plt.ylabel('Absolute Frequency',fontsize=14)
plt.xlabel('cur_ratio/pre_ratio',fontsize=14)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
# Add title
plt.title('ABSOLUTE FREQUENCY HISTOGRAM',fontsize=16);
```

### Add different colors for snap ratio

```python
std = df.snap_change_ratio.std()
df['snap_ratio_flag'] = ''

for i in range(len(df)):
    if df.iloc[i, -2] >= (std*2):
        df.iloc[i, -1] = 'Red'
    elif (df.iloc[i, -2] >= std) and (df.iloc[i, -2] < std*2):
        df.iloc[i, -1] = 'Yellow'
    else:
        df.iloc[i, -1] = 'Green'
```

# Build the Model

#### 1. check the correlation

```python
df.corr()['SNAP_Applications']
```

```python
X = df.drop(columns=['SNAP_Applications'])
y = df.SNAP_Applications
```

#### 2. Result of the linear regression model

```python
import statsmodels.api as sm
import pandas as pd

# Add constant term to X
X = sm.add_constant(X)

# Fit linear regression model
model = sm.OLS(y, X).fit()

# Print summary of regression results
print(model.summary())
```
