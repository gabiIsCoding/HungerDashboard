### Create date_time, last_date_time, last_snap, last_number of workder, last # of disater for the df1

```python



# create date_time
from datetime import datetime
df1['date_time'] = 0

for i in range(len(df1)):
    cur_year = str(df1.iloc[i, 3]) # 3 is the index of year
    cur_mon = str(df1.iloc[i, 4]) # 4 is the index of year
    cur_date_string = cur_year + ' ' + cur_mon
    cur_datetime_string = pd.to_datetime(cur_date_string)
    df1.iloc[i, -1] = cur_datetime_string # -1 is the index of date_time in this step
    

# create last_date_time
df1['last_date_time'] = 0

for i in range(len(df1)):
    dtDate = df1.iloc[i,-2] # -2 is the index of date_time in this step
    df1.iloc[i,-1] = dtDate - pd.DateOffset(months=1)
```

```python
df1.head()
```

### Find last_snap, last_workder, last_disater by the key [last_date_time, County and State]

```python
df1['last_snap'] = None
df1['last_worker'] = None
df1['last_disaster'] = None

for i in range(len(df1)):
    state = df1.iloc[i, 2]
    county = df1.iloc[i, 5]
    last_date_time = df1.last_date_time[i]
    
    filtet_table = df1.loc[(df1.date_time == last_date_time) & (df1.State == state) & (df1.County == county)]
    
    
    if len(filtet_table) != 0:
        df1.iloc[i, -1] = filtet_table['# of disaster'].values[0]
        df1.iloc[i, -2] = filtet_table['Number of Workers'].values[0]
        df1.iloc[i, -3] = filtet_table.SNAP_Applications.values[0]
```
