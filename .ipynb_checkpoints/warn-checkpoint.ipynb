## California

```python
warn_ca.info()
```

```python
warn_ca.head()
```

```python
warn_ca.isna().sum()
```

```python
#drop columns that have NaN == 11534
col_del = ['Temporary/Permanent', 'Union', 'Region', 'Industry', 'Notes', 'Effective Date', 'Layoff/Closure', 'City', 'Company']

warn_ca = warn_ca.drop(columns=col_del)
```

```python
warn_ca.head()
```

```python
warn_ca.isna().sum()
```

```python
# drop rows whose County is na
warn_ca = warn_ca.dropna(axis = 0, subset= 'County')
```

```python
warn_ca['Year'] = warn_ca['Received Date'].dt.year.astype('Int64')
warn_ca['Month'] = warn_ca['Received Date'].dt.month.astype('Int64')
```

```python
sub = ' County'

for idx in range(len(warn_ca)):
    if sub in warn_ca.iloc[idx, 3]:
        warn_ca.iloc[idx, 3] = warn_ca.iloc[idx, 3].replace(sub, '')
```

```python
#filter year > 2018 and year < 2022


warn_ca= warn_ca[(warn_ca.Year > 2018) & (warn_ca.Year < 2022)]
```

```python
warn_ca.drop(columns='Received Date', inplace=True)
```

```python
warn_ca.head()
```

```python
#sum(number of workers) every county every month every year - add column

worker_ca = warn_ca.groupby(['State','Year','Month','County'])['Number of Workers'].sum()
```

```python
worker_ca = worker_ca.to_frame()
```

```python
worker_ca
```

```python
worker_ca = worker_ca.reset_index(level=['State', 'Year', 'Month', 'County'])
```

```python
worker_ca.info()
```
